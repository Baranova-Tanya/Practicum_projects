### Описание проекта

Проект по заказу фермера, владельца молочного хозяйства «Вольный луг». Он хочет купить коров, чтобы расширить поголовье стада. И для этого заключил контракт с ассоциацией пастбищ «ЭкоФерма».
«ЭкоФерма» готова предоставить подробные данные о своих коровах для тщательного отбора перед покупкой.

Фермер определяет качество молока по строгой методике, и при этом ему нужно выполнять свой план развития молочного хозяйства. Он хочет, чтобы каждая корова давала не менее 6000 килограммов молока в год, причем молоко должно быть вкусным (по собственной классификации фермера).

Известно, что технологи фермера планируют для новых коров увеличить значения характеристик корма на 5%.

Разработаем модель машинного обучения, которая поможет фермеру принять объективное решение о покупке. Создадим две прогнозные модели:
* для прогноза возможного удоя коровы (целевой признак `Удой`);
* для расчета вероятности получить вкусное молоко от коровы (целевой признак `Вкус молока`).

С помощью моделей отберем коров по двум критериям:
* средний удой за год — не менее 6000 килограммов;
* молоко вкусное.

### 2.2. Обработка дубликатов

Начнем с явных дубликатов. Проверим их число в каждой таблице.

```python
print(ferma_main.duplicated().sum())
print(cow_buy.duplicated().sum())
print(ferma_dad.duplicated().sum())
```

В таблицах `ferma_main` и `cow_buy` присутствуют явные дубликаты.

Из таблицы `ferma_main` их можно удалить, поскольку в ней каждая корова имеет собственный `id`, и они тоже повторяются. При этом решится проблема несоответствия числа строк в таблицах `ferma_main` и `ferma_dad`, замеченная в шаге 1.

Из таблицы `cow_buy` дубликаты удалять не будем, потому что в ней нет уникальных идентификаторов. Мы не можем быть уверены, что повторяются одни и те же коровы, а не коровы-сестры (причем из-за бинарного отображения возраста неизвестно, совпадает ли их возраст на самом деле), пасущиеся вместе и дающие примерно одинаковое по проценту жира и белка молоко.

```python
ferma_main = ferma_main.drop_duplicates().reset_index(drop=True)
print(ferma_main.duplicated().sum())
```

* `Соверин` и `РефлешнСоверинг` - это линия *Рефлекшн Соверинг* (часто сокращают до *Р. Соверинг*).

Приведем написание названий линий к единообразию.

```python
ferma_main = ferma_main.replace(
    ['Вис Бик Айдиал', 'Айдиал', 'Айдиалл'],
    'Вис Бэк Айдиал'
).replace(
    ['Соверин', 'РефлешнСоверинг'],
    'Рефлекшн Соверинг'
).replace('Равнинные', 'Равнинное')
print('breed:')
print(ferma_main['breed'].value_counts())
print()
print('pasture_type:')
print(ferma_main['pasture_type'].value_counts())
print()
print('dads_breed:')
print(ferma_main['dads_breed'].value_counts())
```

Создадим функцию `calculate_correlation_ratio` для расчета корреляционного отношения и сформируем таблицу, применив ее в цикле к столбцам, содержащим количественные и качественные признаки.

```python
col_num = ['milk_yield',
           'feed_unit',
           'protein_in_feed',
           'sugar_protein_ratio',
           'fat_in_milk',
           'protein_in_milk',
            'sugar_in_feed']

col_cat = ['breed',
           'pasture_type',
           'dads_breed',
           'is_taste_good',
           'is_older',
           'dads_name']

def calculate_correlation_ratio(num_var, cat_var):

    values = np.array(num_var)
    ss_total = np.sum((values.mean() - values) ** 2)

    cats = np.unique(cat_var, return_inverse = True)[1]

    ss_betweengroups = 0

    for c in np.unique(cats):

        group = values[np.argwhere(cats == c).flatten()]
        ss_betweengroups += len(group) * (group.mean() - values.mean()) ** 2

    return round(np.sqrt(ss_betweengroups/ss_total), 2)

corr_r_matrix = pd.DataFrame(columns=col_num)

for col_c in col_cat:
    for col_n in col_num:
        corr_r_matrix.loc[col_c, col_n] = calculate_correlation_ratio(ferma_main[col_n], ferma_main[col_c])

display(corr_r_matrix)
```

Сильных связей нет, есть средние связи между:
* `is_older` и `milk_yield`;
* `is_older` и `feed_unit`;
* `is_older` и `sugar_protein_ratio`;
* `is_older` и `fat_in_milk`.

Слабая связь между `dads_name` и `milk_yield`.

## Шаг 5. Обучение модели линейной регрессии

### 5.1. Модель без учета нелинейности связи `milk_yield` и `sugar_protein_ratio`

Создадим функцию `predict_milk_yield`, которая будет принимать:
* списки столбцов с количественными и категориальными данными для обучения и предсказания;
* метку о том, нужно ли выполнять предсказания для данных из таблицы `cow_buy`.

Функция вернет:
* остатки для тестовой выборки;
* расчитанные метрики *mae*, *mse*, *rmse*, *r_2*;
* предсказания для данных из таблицы `cow_buy` (если аргумент `real_data` будет равен 0, то вместо предсказаний функция вернет 0).

Применение такого рода функции целесообразно, поскольку данных мало, и все обучения будут происходить очень быстро.

```python
def predict_milk_yield(col_num, col_cat, real_data):
    RANDOM_STATE = 42

    X = ferma_main[col_num + col_cat]
    y = ferma_main['milk_yield']
    X_train, X_test, y_train, y_test = train_test_split(
        X,
        y,
        random_state=RANDOM_STATE
    )

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train[col_num])
    X_test_scaled = scaler.transform(X_test[col_num])

    encoder = OneHotEncoder(drop='first', sparse=False)
    X_train_ohe = encoder.fit_transform(X_train[col_cat])
    X_test_ohe = encoder.transform(X_test[col_cat])

    X_train = pd.DataFrame(
        X_train_ohe,
        columns=encoder.get_feature_names()
    ).join(pd.DataFrame(X_train_scaled))

    X_test = pd.DataFrame(
        X_test_ohe,
        columns=encoder.get_feature_names()
    ).join(pd.DataFrame(X_test_scaled))


    model_lr = LinearRegression()
    model_lr.fit(X_train, y_train)

    predictions_test = model_lr.predict(X_test)
    errors = y_test - predictions_test

    metrics = pd.DataFrame.from_dict(
            {'mae': [mean_absolute_error(y_test, predictions_test)],
             'mse': [mean_squared_error(y_test, predictions_test)],
             'rmse': [mean_squared_error(y_test, predictions_test, squared=False)],
             'r_2': [r2_score(y_test, predictions_test)]
            },
             orient='index')

    if real_data == 0:
        predictions_real = 0

    else:
        X_real = cow_buy[col_num + col_cat]
        X_real_scaled = scaler.transform(X_real[col_num])
        X_real_ohe = encoder.transform(X_real[col_cat])
        X_real = pd.DataFrame(
            X_real_ohe,
            columns=encoder.get_feature_names()
        ).join(pd.DataFrame(X_real_scaled))

        predictions_real = pd.Series(model_lr.predict(X_real))


    return errors, metrics, predictions_test, predictions_real
```

Сначала обучим модель на следующих данных:
* `sugar_protein_ratio` (сильная линейная связь с целевым признаком)
* `feed_unit` (сильная линейная связь с целевым признаком)
* `fat_in_milk` (средняя линейная связь с целевым признаком)
* `is_older` (средняя связь с целевым признаком)
* `sugar_in_feed` (использовать одновременно `sugar_in_feed` и `protein_in_feed` не следует из-за их общей прямой взаимосвязи с `sugar_protein_ratio`, при этом `sugar_in_feed` теснее связан с целевым признаком)

Они никак не могут находиться в причинно-следственной связи, но оба эти признака зависят от того, насколько корова в целом в хорошей форме. То есть жирность молока может рассматриваться как косвенный показатель общего благополучия коровы, которое несомненно влияет на удой. А прямых показателей благополучия коровы (информация о том, не болела ли корова, не было ли стрессов, как конкретно она питается и сколько пьет воды) у нас нет.

## Шаг 6. Обучение модели логистической регрессии

### 6.1. Подбор входных признаков и обучение модели

Вкус молока слабо связан с остальными признаками, как показал корреляционный анализ. Поэтому начнем с того, что используем для обучения модели все имеющиеся признаки.

```python
RANDOM_STATE = 42

X = ferma_main[['is_older', 'sugar_in_feed', 'feed_unit', 's_p_ratio_bi', 'fat_in_milk', 'protein_in_milk',
                    'pasture_type', 'dads_name', 'breed']]
y = ferma_main['is_taste_good']
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    random_state=RANDOM_STATE
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train[[
    'sugar_in_feed', 'feed_unit', 'fat_in_milk', 'protein_in_milk']])
X_test_scaled = scaler.transform(X_test[[
    'sugar_in_feed', 'feed_unit', 'fat_in_milk', 'protein_in_milk']])

encoder = OneHotEncoder(drop='first', sparse=False)
X_train_ohe = encoder.fit_transform(X_train[['pasture_type', 'dads_name', 'breed', 'is_older', 's_p_ratio_bi']])
X_test_ohe = encoder.transform(X_test[['pasture_type', 'dads_name', 'breed', 'is_older', 's_p_ratio_bi']])

X_train = pd.DataFrame(
    X_train_ohe,
    columns=encoder.get_feature_names()
).join(pd.DataFrame(X_train_scaled))

X_test = pd.DataFrame(
    X_test_ohe,
    columns=encoder.get_feature_names()
).join(pd.DataFrame(X_test_scaled))


clf = LogisticRegression(class_weight='balanced')
clf.fit(X_train, y_train)

predictions_test = clf.predict(X_test)

confusion_matrix_log = confusion_matrix(y_test, predictions_test)

metrics_log = pd.DataFrame.from_dict(
        {'accuracy': [accuracy_score(y_test, predictions_test)],
        'precision': [precision_score(y_test, predictions_test)],
        'recall': [recall_score(y_test, predictions_test)]
        },
        orient='index')


display(confusion_matrix_log)
display(metrics_log)
display(pd.DataFrame(clf.predict_proba(X_test)).describe())
```

Для фермера критичны ошибки первого рода `FalsePositive`, когда невкусное молоко модель посчитала вкусным.

Метрика `precision` (точность) в данном случае важнее метрики `recall` (полнота).

Получившаяся модель уже справляется лучше простого угадывания (`accuracy` ощутимо больше 0.5), однако нужно добиться того, чтобы ошибок первого рода не было совсем, а `precision` была максимальной.

По вероятностям отношения к тому или иному классу видно, что модель часто "не уверена": большое число вероятностей близки к 0.5 (при хорошо работающей модели они должны стремиться к границам промежутка от 0 до 1).

Скорее всего, точность модели возрастет, если исключить параметры, которые связаны со вкусом молока еще меньше, чем остальные.

Дополнительно транформируем содержание жира в молоке в бинарный признак (эту возможность мы рассмотрели в пункте 4.2.4)

Кроме того, изучение информации о вкусе молока показало, что низкая продуктивность коровы может свидетельствовать о наличии заболеваний, отрицательно сказывающихся на вкусе молока. Поэтому попробуем добавить признак `milk_yield`.

```python
ferma_main['fat_in_milk_bi'] = np.where(ferma_main['fat_in_milk']>3.69, 1, 0)
ferma_main['fat_protein_ratio'] = ferma_main['fat_in_milk'] / ferma_main['protein_in_milk']

display(ferma_main.head())

print(ferma_main['fat_in_milk_bi'].corr(ferma_main['is_taste_good'], method='spearman'))
print(ferma_main['fat_protein_ratio'].corr(ferma_main['is_taste_good'], method='spearman'))
```

```python
X = ferma_main[['feed_unit', 'is_older', 'sugar_in_feed', 's_p_ratio_bi', 'fat_in_milk_bi', 'protein_in_milk',
                    'pasture_type', 'dads_name', 'fat_protein_ratio', 'milk_yield']]
y = ferma_main['is_taste_good']
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    random_state=RANDOM_STATE
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train[[
    'feed_unit', 'sugar_in_feed', 'protein_in_milk', 'fat_protein_ratio', 'milk_yield']])

X_test_scaled = scaler.transform(X_test[[
    'feed_unit', 'sugar_in_feed', 'protein_in_milk', 'fat_protein_ratio', 'milk_yield']])

encoder = OneHotEncoder(drop='first', sparse=False)
X_train_ohe = encoder.fit_transform(X_train[['pasture_type', 'is_older', 'dads_name', 's_p_ratio_bi']])
X_test_ohe = encoder.transform(X_test[['pasture_type', 'is_older', 'dads_name', 's_p_ratio_bi']])

X_train = pd.DataFrame(
    X_train_ohe,
    columns=encoder.get_feature_names()
).join(pd.DataFrame(X_train_scaled))

X_test = pd.DataFrame(
    X_test_ohe,
    columns=encoder.get_feature_names()
).join(pd.DataFrame(X_test_scaled))


clf = LogisticRegression(class_weight='balanced')
clf.fit(X_train, y_train)

predictions_test = clf.predict(X_test)

confusion_matrix_log = confusion_matrix(y_test, predictions_test)

metrics_log = pd.DataFrame.from_dict(
        {'accuracy': [accuracy_score(y_test, predictions_test)],
        'precision': [precision_score(y_test, predictions_test)],
        'recall': [recall_score(y_test, predictions_test)]
        },
        orient='index')


display(confusion_matrix_log)
display(metrics_log)
display(pd.DataFrame(clf.predict_proba(X_test)).describe())
```

Изменения позволили сократить число ошибок первого рода.

Попробуем подобрать пороговое значение, при котором такие ошибки будут исключены.

### 6.2. Подбор порогового значения для исключения ошибок первого рода

```python
y_proba = clf.predict_proba(X_test)[:, 1]
data = pd.DataFrame()
data['y_proba'] = y_proba
for i in range(1, 50):
    t = 0.5 + i / 100

    data['y_pred_'+str(t)] = data['y_proba'].apply(lambda x: 1 if x > t else 0)
    errors_FP = confusion_matrix(y_test, data['y_pred_'+str(t)])[0, 1]
    if errors_FP == 0:
        print('Порог классификации:', t)
        display(confusion_matrix(y_test, data['y_pred_'+str(t)]))
        display(pd.DataFrame(data=[
            accuracy_score(y_test, data['y_pred_'+str(t)]),
            precision_score(y_test, data['y_pred_'+str(t)]),
            recall_score(y_test, data['y_pred_'+str(t)])],
            index=['accuracy', 'precision', 'recall']))
        break
```

Мы добились полного исключения ошибок первого рода, однако теперь модель слишком плохо распознает вкусное молоко. Только 5% коров, дающих вкусное молоко, распознаны моделью, а число ошибок второго рода чрезмерно выросло.

### 7.2. Краткие рекомендации для фермера

* однозначно к покупке можно рекомендовать двух коров, поскольку только для них мы уверенно предсказываем хороший вкус молока, а достаточные удои – для всех 20 коров;
* чтобы взвешенно принять решение о покупке прочих коров (если это необходимо), в таблице есть столбец `y_proba`, который содержит нашу оценку вероятности того, что корова дает вкусное молоко;
* чтобы сделать работу модели более точной, нужно больше данных, связанных со вкусом молока (в предоставленных данных почти никакие показатели не имеют значимого влияния на вкус молока, поэтому прогнозы такие несовершенные);
* это могли бы быть данные о:
  - содержании лактозы в молоке (это один из ключевых параметров, определяющих вкус);
  - содержании солей в молоке;
  - семействе коровы (наследственность по материнской линии влияет на вкусовые качества молока);
  - употребляемой воде;
  - годе рождения коровы или числе лактаций (это влияет и на удой, и на вкус молока).
