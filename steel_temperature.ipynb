```python
def show_num(data, col):
    """
    Принимает:
        - наименование переменной, хранящей датафрейм;
        - наименование столбца с количественными данными в этом датафрейме.
    Возвращает:
        - общее описание данных;
        - гистограмму с наложенным графиком плотности распределения;
        - диаграмму размаха.
    """
    figure, axs = plt.subplots(1, 2, figsize=(15, 8))
    figure.suptitle(col.title())

    sns.histplot(data[col], bins=50, kde=True, ax=axs[0])
    axs[0].set_title('Гистограмма и плотность распределения')

    sns.boxplot(data=data[col], ax=axs[1])
    axs[1].set_title('Диаграмма размаха')

    plt.show()

    display(data[col].describe().to_frame())

def rename_cols(df):
    """
    Принимает:
        - наименование переменной, хранящей датафрейм.
    Изеняет датафрейм:
        - приводит названия столбцов к нижнему регистру;
        - заменяет пробелы в названиях столбцов на нижнее подчеркивание.
    """
    new_cols = []
    for col in df.columns:
        new_col = col.lower().replace(' ', '_')
        new_cols.append(new_col)
    df.columns = new_cols

def del_useless(df, n):
    """
    Принимает:
        - наименование переменной, хранящей датафрейм;
        - наименьшее число значений в столбце для его сохранения.
    Возвращает:
        - датафрейм, из которого удалены столбцы, в которых количество значений меньше указанного числа;
        - список удаленных столбцов.
    """
    df_cols = df.columns
    df_cols.drop('key')

    useless_df_cols = []

    for col in df_cols:
        not_nans = df[col].notna().sum()
        if not_nans < n:
            useless_df_cols.append(col)

    df_use = df.drop(useless_df_cols, axis=1)
    return df_use, useless_df_cols
```

### `bulk_time`

Проверим, совпадают ли партии в этой таблице с партиями из таблицы `bulk`.

```python
print('Количество совпадающих партий:',
      (bulk['key'].sort_values()==bulk_time['key'].sort_values()).sum())
```

Обе таблицы действительно содержат данные об одних и тех же партиях.

Проверим, относятся ли они к тому же промежутку времени, что и данные об электродах.

```python
years = []
months = []
days = []

cols_bulk = bulk_time.columns.tolist()
cols_bulk.remove('key')

for col in cols_bulk:
    year = pd.DatetimeIndex(bulk_time[col]).year.value_counts().sort_index().index.astype('str')
    year = ', '.join(year).replace('.0', '')
    years.append(year)

    month = pd.DatetimeIndex(bulk_time[col]).month.value_counts().sort_index().index.astype('str')
    month = ', '.join(month).replace('.0', '')
    months.append(month)

    day = pd.DatetimeIndex(bulk_time[col]).day.value_counts().sort_index().index.astype('str')
    day = ', '.join(day).replace('.0', '')
    days.append(day)

df = pd.DataFrame({'столбцы': cols_bulk,
                 'годы': years,
                 'месяцы': months,
                 'дни': days})
display(df)
```

Все совпадает: данные о сыпучих материалах относятся к тому же временному промежутку, что и данные об электродах.

```python
msno.heatmap(bulk_time)
plt.title('Матрица корреляции пропусков')
plt.show()
```

Пропуски в таблице `bulk_time` взаимосвязаны точно так же, как и в таблице `bulk`.

Проверим порядок, в котором указаны данные о времени в столбцах.

```python
columns = []
exist = []
revers = []



for i in range(1, 15):
    col_1 = 'Bulk ' + str(i)
    col_2 = 'Bulk ' + str(i + 1)

    data = bulk_time.loc[(bulk_time[col_1].notnull()==True) & (bulk_time[col_2].notnull()==True)][[col_1, col_2]]

    columns.append(col_1 + ' и ' + col_2)
    exist.append(data.shape[0])
    revers.append((data[col_1]>data[col_2]).sum())


df = pd.DataFrame({'столбцы': columns,
                 'есть данные в обоих': exist,
                 'обратный порядок': revers})
display(df)
```

Данные о времени подачи сыпучих материалов не упорядочены. Есть сомнения, что время и объем, указанные в столбцах с одинаковыми номерами, действительно соответствуют друг другу. Неизвестно, нарушается ли порядок измерений в таблице `bulk`.
Выделим из них минимум и максимум для каждой партии, чтобы позже сравнить с другими имеющимися данными о времени.

```python
bulk_time_use = bulk_time.drop(cols_bulk, axis=1)
bulk_time_use['bulk_time_min'] = list(bulk_time[cols_bulk].agg('min', axis=1))
bulk_time_use['bulk_time_max'] = list(bulk_time[cols_bulk].agg('max', axis=1))
display(bulk_time_use.head(5))
```

**Промежуточные выводы:**
1. Данные о сыпучих материалах относятся к тому же периоду, что и данные об электродах. Значит, можно будет соединять таблицы по номерам партий.
2. Использовать для обучения данные из таблицы `bulk_time` крайне затруднительно.
3. Сформировали таблицу `bulk_time_use`, содержащую минимальное и максимальное время из таблицы `bulk_time` для каждой партии.

```python
years = []
months = []
days = []

cols_wire = wire_time.columns.tolist()
cols_wire.remove('key')

for col in cols_wire:
    year = pd.DatetimeIndex(wire_time[col]).year.value_counts().sort_index().index.astype('str')
    year = ', '.join(year).replace('.0', '')
    years.append(year)

    month = pd.DatetimeIndex(wire_time[col]).month.value_counts().sort_index().index.astype('str')
    month = ', '.join(month).replace('.0', '')
    months.append(month)

    day = pd.DatetimeIndex(wire_time[col]).day.value_counts().sort_index().index.astype('str')
    day = ', '.join(day).replace('.0', '')
    days.append(day)

df = pd.DataFrame({'столбцы': cols_wire,
                 'годы': years,
                 'месяцы': months,
                 'дни': days})
display(df)
```

### `temp`. Группировка и агрегирование

Пронумеруем измерения температуры, чтобы разобраться, сколько всего их было.

Затем создадим сводную таблицу, сгруппировав данные по номеру партии, где отразим первое и последнее измерение (температуру и время), а также среднюю температуру для всех промежуточных измерений.


```python
temp = temp.sort_values(by='time')
temp.loc[:,'cum_count'] = temp.groupby('key').cumcount() + 1
display(temp.head(10))
display(temp.info())
temp['cum_count'].value_counts()
```

```python
max_counts = temp.groupby('key').agg({'cum_count' : 'max'})
max_counts.value_counts()
```

Вo многих партиях ограничились единственным замером температуры. Такие партии, скорее всего, не подходят для обучения: в них должен отсутствовать целевой признак. Чтобы выяснить это достоверно, нам нужно будет после объединения таблиц сравнить время последнего замера температуры с другими известными данными о времени.

```python
temp_use = temp

temp_use['first_measure_time'] = temp_use['time']
temp_use['last_measure_time'] = temp_use['time']
temp_use['first_measure_temp'] = temp_use['temperature']
temp_use['last_measure_temp'] = temp_use['temperature']
temp_use['max_counts'] = temp_use['cum_count']

temp_use = temp_use.groupby('key').agg({
    'first_measure_time' : 'first',
    'last_measure_time' : 'last',
    'first_measure_temp' : 'first',
    'last_measure_temp' : 'last',
    'max_counts': 'max'
}).reset_index()

display(temp_use.head(10))
```

Группировка прошла нормально. Рассчитаем среднюю температуру для промежуточных измерений.

```python
temp = temp[['key',
             'time',
             'temperature',
             'cum_count']].merge(temp_use[['key', 'max_counts']],
                                 on='key',
                                 how='left')
temp_rest = temp.loc[
    (temp['cum_count']!=temp['max_counts'])]\
.groupby('key').agg(
    {'temperature' : 'mean'}
).reset_index()

temp_rest.describe()
```

```python
temp_use['rest_mean_temp'] = temp_rest['temperature']
print(temp_use['rest_mean_temp'].describe())
temp_use.info()
```

### `arc`. Группировка и агрегирование

Убедимся, что время конца нагрева везде больше времени начала.

Пронумеруем итерации нагрева и сравним их с количеством замеров температуры. Если процесс налажен хорошо, измерений температуры в большинстве партий будет по крайней мере на 1 больше, чем итераций нагрева.

```python
normal_time = (arc['end_heating'] > arc['start_heating']).sum()
rows = arc.shape[0]
if normal_time == rows:
    print('Временных парадоксов нет')
elif normal_time < rows:
    print('Обнаружены временные парадоксы')
    print(f'в {rows-normal_time} строках')
else:
    print('Ошибка')
```

```python
arc.info()
```

```python
arc = arc.sort_values(by='start_heating')
arc.loc[:,'cum_count'] = arc.groupby('key').cumcount() + 1
arc.head(10)
```

```python
max_counts_arc = arc.groupby('key').agg({'cum_count' : 'max'})
max_counts_arc.value_counts()
```

Число итераций нагрева мало связано с числом замеров температуры. Скорее всего, акцентировать внимание на отдельных операциях бессмысленно, поскольку не понятно, насколько упорядоченно велся сбор данных.

Рассчитаем полную мощность для каждой партии и совокупное время нагревания. Отметим время начала первого нагревания и окончания последнего, чтобы позже сравнить со временем замеров температуры.


```python
arc['full_power'] = np.sqrt(arc['active_power']**2 + arc['reactive_power']**2)
arc['time_heating'] = (arc['end_heating'] - arc['start_heating'])
arc['time_heating'] = arc['time_heating'].apply(lambda x: x.seconds)

arc.head()
```

```python
arc_use = arc.drop(['active_power', 'reactive_power'], axis=1)
arc_use.columns
```

```python
arc_use = arc_use.groupby('key').agg({
    'start_heating' : 'first',
    'end_heating' : 'last',
    'cum_count' : 'max',
    'full_power' : 'sum',
    'time_heating' : 'sum'
}).reset_index()

arc_use.columns = ['key',
                   'first_start_heating',
                   'last_end_heating',
                   'max_cum_count',
                   'sum_full_power',
                   'sum_time_heating']
display(arc_use.head(10))
```

Группировка прошла нормально.

Теперь все таблицы можно объединить по номеру партии.

### Объединение таблиц

```python
data = arc_use.merge(
    bulk_use, how='outer', on='key'
).merge(
    bulk_time_use, how='outer', on='key'
).merge(
    wire_use, how='outer', on='key'
).merge(
    wire_time_use, how='outer', on='key'
).merge(
    gas, how='outer', on='key'
).merge(
    temp_use, how='outer', on='key'
)

display(data.head())
data.info()
```

Сравним время последних замеров температуры со временем последнего окончания нагрева, последних подач сыпучих и проволочных материалов. Посмотрим, сколько партий имеют целевой признак на самом деле.

```python
data['time_compare_1'] = (
    data['last_measure_time'] >= data['last_end_heating']
) & (
    data['last_measure_time'] >= data['bulk_time_max']
) & (
    data['last_measure_time'] >= data['wire_time_max']
)
print('Последний замер температуры есть:', data['time_compare_1'].sum())

data['time_compare_2'] = (
    data['first_measure_time'] <= data['first_start_heating']
) & (
    data['first_measure_time'] <= data['bulk_time_min']
) & (
    data['first_measure_time'] <= data['wire_time_min']
)
print('Первый замер температуры есть:', data['time_compare_2'].sum())

data['time_compare'] = data['time_compare_1'] & data['time_compare_2']
print('Первый и последний замеры температуры есть:', data['time_compare'].sum())
```

Данные 2327 строк могут быть использованы для обучения. Остальные строки удалим. Удалим также столбцы, которые мы использовали для выявления первого и последнего замеров и прочих расчетов.

```python
data = data.loc[data['time_compare']==True].drop([
    'time_compare_1',
    'time_compare_2',
    'time_compare',
    'first_start_heating',
    'last_end_heating',
    'bulk_time_min',
    'bulk_time_max',
    'wire_time_min',
    'wire_time_max',
    'first_measure_time',
    'last_measure_time',
    'max_counts'
], axis=1)
data.info()
```

## Шаг 6. Обучение моделей машинного обучения

Масштабируем входные признаки двумя способами:
- `StandardScaler()`;
- `MinMaxScaler()`.

Обучим три модели:
- `LinearRegression()`;
- `SVR()` (с вариантами гиперпараметров);
- `LGBMRegressor()` (с вариантами гиперпараметров).

Составим для этого пайплайн.

```python
data_preprocessor = ColumnTransformer(
    [('preprocessor', StandardScaler(), X_train.columns)],
    remainder='passthrough'
)

pipe = Pipeline(
    [('preprocessor', data_preprocessor),
     ('models', LinearRegression())]
)

params = [
    {
        'models': [LinearRegression()],
        'preprocessor': [StandardScaler(), MinMaxScaler()]
    },
    {
        'models': [SVR()],
        'models__kernel': ['linear', 'rbf', 'sigmoid'],
        'preprocessor': [StandardScaler(), MinMaxScaler()]
    },
    {
        'models': [LGBMRegressor(random_state=RANDOM_STATE)],
        'models__boosting_type': ['gbdt', 'dart'],
        'preprocessor': [StandardScaler(), MinMaxScaler()]

    }
]
```

### Выводы по шагу 6
1. Составили пайплайн для масштабирования данных и обучения трех разных регрессионных моделей с подбором гиперпараметров.

## Шаг 7. Выбор лучшей модели

### Поиск лучшей модели
Выберем лучшую модель при помощи `RandomizedSearchCV`, поскольку перебираем много гиперпараметров.

Оценивать модели будем по средней абсолютной ошибке.

```python
warnings.simplefilter(action='ignore', category=FutureWarning)
search = RandomizedSearchCV(
        pipe,
        params,
        scoring='neg_mean_absolute_error',
        n_iter=10,
        n_jobs=-1,
        random_state=RANDOM_STATE,
        verbose=2
)
search.fit(X_train_imputed, y_train)

print('Лучшая модель и ее параметры:\n', search.best_params_)
print()
print ('Метрика MAE на кроссвалидации:', search.best_score_ * -1)
```

```python
results = pd.DataFrame(search.cv_results_)
results[['param_models',
         'params',
         'mean_fit_time',
         'mean_test_score',
         'rank_test_score']]
```

Лучшей моделью на кроссвалидации оказалась `LGBMRegressor()` с типом бустинга `Gradient Boosting Decision Tree`. Масштабирование произведено при помощи `MinMaxScaler()`.

Метрика лучшей модели преодолевает необходимое пороговое значение. Метрики других моделей в основном не сильно отличаются (значительно хуже справилась только `LGBMRegressor()` с ядром `Dropouts meet Multiple Additive Regression Trees`).

### Проверка лучшей модели на тестовой выборке


```python
warnings.simplefilter(action='ignore', category=FutureWarning)
best_model = search.best_estimator_.named_steps['models']
best_preprocessor = search.best_estimator_.named_steps['preprocessor']

y_pred = best_model.predict(best_preprocessor.transform(X_test_imputed))
print ('Метрика MAE на тестовой выборке:',
       mean_absolute_error(y_test, y_pred))
```

На тестовой выборке также получили удовлетворительное качество.
